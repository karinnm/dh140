```python
characters = [
    'Prince',
    'Par',
    'Mon',
    'Cap',
    'Rom',
    'Tyb',
    'Mer',
    'Ben',
    'Friar',
    'Laur',
    'John',
    'Bal',
    'Abr',
    'Samp',
    'Greg',
    'Peter',
    'Apoth',
    'Mus',
    'Officer',
    'M. Wife',
    'Lady',
    'Jul',
    'Nurse',
    'Citizen'
]
```

```python
len(acts)
```

```python
charnum = {}

connections = {}
for i in range(len(characters)-1):
    for j in range(i+1,len(characters)):
        connections[(characters[i],characters[j])] = 0

for k in characters:
    charnum[k] = 0

for i in acts:
    for j in i.split('Scene')[1:]:
        scenechars = []
        for k in characters:
            if j.find(k) != -1:
                scenechars.append(k)
                charnum[k] += 1
        for a in range(len(scenechars)-1):
            for b in range(a+1,len(scenechars)):
                connections[(scenechars[a],scenechars[b])] += 1
```

#### The graph contains information on characters and their connections. Each node represents a character in the play and the size of the node represents how many scenes that character is in. The edges represent the connections between characters, so characters that are connected by an edge occur and speak in a scene together. The width of each edge is the weight of the connection between those two characters relative to the max value of all the weights. So, the thicker the edge is, the more scenes these characters occur in together and the more interaction between them. 

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
nxgraphitems = []

for k,v in connections.items():
    if v != 0:
        dashedge = (k[0],k[1])
        nxgraphitems.append(dashedge)
    
nxgraphitems
```

```python
G = nx.Graph()
G.add_edges_from(nxgraphitems)
```

```python
G.nodes
```

```python
nx.degree_centrality(G)
```

```python
dc = nx.degree_centrality(G)
for i in sorted(dc, key=dc.get, reverse=True):
    print('{:15s}: {:.3f}'.format(i.title(), dc[i]))
```

```python
list(G.neighbors('Rom'))
```

```python
nx.shortest_path(G,'Rom','Jul')
```

```python
nx.betweenness_centrality(G)
```

```python
dc = nx.betweenness_centrality(G)
for i in sorted(dc, key=dc.get, reverse=True):
    print('{:15s}: {:.3f}'.format(i.title(), dc[i]))
```

```python
from networkx.algorithms.community.centrality import girvan_newman
```

```python
communities = girvan_newman(G)
```

```python
node_groups = []
for com in next(communities):
    node_groups.append(list(com))
 
print(node_groups)
 
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append('blue')
    else:
        color_map.append('green')
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

```python
nx.density(G)
```

```python
graphitems = []

for k,v in charnum.items():
    if k in node_groups[0]:
        modularity = 'blue'
    elif k in node_groups[1]:
        modularity = 'green'
    else:
        modularity = 'red'
    dashnode = {'data': {'id': k,
                         'label': k.title(),
                         'size': str(v),'modularity': modularity}}
    graphitems.append(dashnode)

correction_factor = max(connections.values())
print(correction_factor)
for k,v in connections.items():
    if v != 0:
        dashedge = {'data': {'source': k[0],
                             'target': k[1],
                             'weight': v/correction_factor}}
        graphitems.append(dashedge)
    
graphitems
```

```python
app = JupyterDash(__name__)

app.layout = html.Div([
    cyto.Cytoscape(
        layout={'name': 'cose'},
        elements=graphitems,
        style={'width': '100%', 'height': '600px'},
        stylesheet=[
            {
                'selector': 'node',
                'style': {
                    'content':'data(label)',
                    'text-halign':'center',
                    'text-valign':'center',
                    'width': 'data(size)',
                    'height': 'data(size)',
                    'font-size':6,
                    'color': 'white',
                    'background-color': 'blue',
                    'text-outline-color': 'black',
                    'text-outline-width': 0.5,
                    'shape':'circle'
                }
            },
            {
                'selector':'edge',
                'style': {
                    'width':'data(weight)',
                    'line-color': 'gold',
                  }
            },
        ]
    )
])

app.run_server(debug=True)
```
