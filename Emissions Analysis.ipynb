```python
# creating a new function definition that returns the country with the highest emissions per GDP in a given year
```

```python
def worst_emitter_of_given_year(dataset, year):
    # Part 1:
    # create empty list 
    year_entries = []
    # for loop to get all countries emission information in year of interest
    for entry in emissions:
        if entry['Year'] == year:
            year_entries.append(entry)
    # Part 2:
    # set maximum variable at 0 to start
    maximum_emissions = 0
    for entry in year_entries:
        ratio_per_GDP = entry['Ratio']['Per GDP']
        country = entry['Country']
        # conditional expression that finds maximum emissions per GDP
        if ratio_per_GDP > maximum_emissions:
            maximum_emissions = ratio_per_GDP
            max_country = country
    final_output = "The country with the highest emissions per GDP is " + str(max_country) + " with per_gdp emissions of: " + str(maximum_emissions)
    return final_output
```

```python
# running the function for the year 1980
```

```python
worst_emitter_of_given_year(emissions, 1980)
```

```python
# running the function for the year 2000
```

```python
worst_emitter_of_given_year(emissions, 2000)
```

```python
# Qatar has the highest emissions per GDP in both 1980 and 2000
```
