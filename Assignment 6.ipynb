## Step 2: Shakespeare Word Frequency

```python
import requests
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
# importing Shakespeare's play Macbeth from Project Gutenberg 
```

```python
response = requests.get('https://www.gutenberg.org/cache/epub/2264/pg2264.html')
```

```python
response
```

```python
from bs4 import BeautifulSoup
```

```python
document = BeautifulSoup(response.text, "html.parser")
```

```python
type(document)
```

```python
text = document.text
```

```python
sent = sent_tokenize(text)
print(sent)
```

```python
# finding where the play actually begins
```

```python
sent[140]
```

```python
# finding where the play ends
```

```python
sent[1603]
```

```python
new_sent = sent[140:1603]
```

```python
words = []
for s in new_sent:
    for w in word_tokenize(s):
        words.append(w)
```

```python
print(words)
```

```python
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
print(punctuation)
```

```python
print(stopwords.words('english'))
```

```python
print(list(punctuation))
```

```python
myStopWords = list(punctuation) + stopwords.words('english')
```

```python
print(myStopWords)
```

```python
wordsNoStop = []
for i in words:
    if i not in myStopWords:
        wordsNoStop.append(i)
print(words)
print(wordsNoStop)
```

```python
wordsNoStopComp = [w for w in words if w not in myStopWords]
print(wordsNoStopComp)
```

```python
from nltk.probability import FreqDist
```

```python
# finding the 20 most frequent words
```

```python
freq = FreqDist(wordsNoStopComp)
```

```python
freq
```

```python
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

```python
from nltk.stem.porter import PorterStemmer
#from nltk.stem import LancasterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
# stemming the words
```

```python
listOfStemmedWords = []
for i in new_movie_reviews:
    listOfStemmedWords.append([p_stemmer.stem(w) for w in i])
```

```python
listOfStemmedWords[0]
```

```python
!pip install gensim
```

```python
from gensim import corpora, models
import gensim
```

```python
dictionary = corpora.Dictionary(listOfStemmedWords)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in listOfStemmedWords]
```

```python
print(corpus[5])
```

```python
print(dictionary.token2id['drama'])
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=2, 
                                           id2word = dictionary, 
                                           passes=20)
```

```python
for i in ldamodel.print_topics(num_topics=2, num_words=20):
    print(i)
```
